```python
import numpy as np
import matplotlib.pyplot as plt

# Function Definitions
def buy_call(S, K, P):
    return np.maximum(0, S - K) - P

def sell_call(S, K, P):
    return P - np.maximum(0, S - K)

def buy_put(S, K, P):
    return np.maximum(0, K - S) - P

def sell_put(S, K, P):
    return P - np.maximum(0, K - S)

# Input Parameters
K = float(input("Enter Strike Price (K): "))
P = float(input("Enter Premium (P): "))
spread = 50  # You can change this range if needed
S = np.linspace(K - spread, K + spread, 100)

# Calculating Payoffs
payoff_buy_call = buy_call(S, K, P)
payoff_sell_call = sell_call(S, K, P)
payoff_buy_put = buy_put(S, K, P)
payoff_sell_put = sell_put(S, K, P)

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(S, payoff_buy_call, label='Buy Call', color='green')
plt.plot(S, payoff_sell_call, label='Sell Call', color='red')
plt.plot(S, payoff_buy_put, label='Buy Put', color='blue')
plt.plot(S, payoff_sell_put, label='Sell Put', color='orange')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(K, color='grey', linestyle='--', label='Strike Price (K)')
plt.title('Option Payoffs: Profit/Loss vs Spot Price')
plt.xlabel('Spot Price at Expiry (S)')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)
plt.show()
```

The strike price is constant at 100
The buy call and sell call profit is 0 at 120(strike price + premium) ie. breakeven occurs at 120.
The sell call was constant till 100 and then started to monotonically decrease
The buy call was constant till 100 and then started to monotonically increase.
The buy put and sell put profit is 0 at 80(strike price - premium) ie. breakeven occurs at 80.
The buy put was monotonically decreasing till 100 and then became constant.
The sell put was monotonically increasing till 100 and then became constant.
